```python
import cv2
import numpy as np
import matplotlib.pyplot as mpl

photo_1 = cv2.imread("photo-1.jpg")
photo_2 = cv2.imread("photo-2.jpg")

photo_3 = cv2.imread("photo-3.jpeg")
```

```python
mpl.imshow(photo_2)
```

```python
mpl.imshow(photo_3)
```

```python
from PIL import Image             #Resizing the photo
image = Image.open("photo-2.jpg")
photo_2 = image.resize((500,625))
photo_2.save("photo-2_resize.jpg")
photo_2_new = cv2.imread("photo-2_resize.jpg")
```

```python
mpl.imshow(photo_2_new)
```

```python
photo_collage = np.hstack((photo_2_new, photo_3))    #Attaching 2 images horizontolly
```

```python
mpl.imshow(photo_collage)
```

```python
photo_collage.shape
```

```python
photo_collage_new = np.vstack((photo_1, photo_collage))     #Attaching previous collaged image with another image vertically
```

```python
mpl.imshow(photo_collage_new)
```

```python
photo_collage_new.shape
```

```python
cv2.imwrite("photo_collage_new.jpg", photo_collage_new)
```

```python
from PIL import Image                     #Resizing the image to look good
image = Image.open("photo_collage_new.jpg")
photo_new_collage_2 = image.resize((2000, 2116))
photo_new_collage_2.save("photo_collage_new_2.jpg")
photo_new_collage_2 = cv2.imread("photo_collage_new_2.jpg")
```

```python
mpl.imshow(photo_new_collage_2)   #Final Collaged image
```
